#2. Check if a number is even or odd.

```python
num = 21
if num % 2 == 0:
  print(f'{num} is Even')
else:
  print(f'{num} is Odd')
```

#3. Check if a number is divisible by 5.

```python
num = 20
if num % 2 == 5:
  print(f'{num} is divisible by 5')
else:
  print(f'{num} is not divisible by 5')
```

#4. Check if a number is divisible by both 3 and 5

```python
num = 15
if num % 5 == 0 and num % 3 == 0:
  print(f'{num} is divisible with 3 & 5')
else:
  print(f'{num} is not divisible with 3 & 5')
```

#5. Check if a given year is a leap year

```python
year = 2024
if year % 4 == 0 and (not year % 100 == 0 or year % 400 == 0):
  print(f'{year} is a leap year')
else:
  print(f'{year} is not leap year')
```

# 8. Take a temperature value and print "Cold", "Warm", or "Hot" using coditions

```python
given_temp = 15
if given_temp < 20:
  print("cold")
elif 20 <= given_temp <= 30:
  print("Warm")
else:
  print("Hot")
```

#9. Take a character and check if it's a vowel or constant

```python
vowels = ['a', 'e', 'i', 'o', 'u']
char = 'A'
if char.lower() in vowels:
  print('Vowel')
else:
  print('Constant')
```

#10. Take a character and check whether it's uppercase, lowercase, a digit, or a special character

```python
char = input("Enter a character: ")

if char.islower():
  print('Lower case')
elif char.isupper():
  print('Upper case')
elif char.isdigit():
  print('Digit')
else:
  print('Special Character')
```
