#1. Take three sides adn check if they form a valid triangle

```python
side1 = input('Enter the side1: ')
side2 = input('Enter the side2: ')
side3 = input('Enter the side3: ')

if side1 + side2 > side3 and side2 + side1 > side3 and side3 + side1 > side2:
  print("Valid Triangle")
else:
  print("Not a valid triangle")
```

#2. if the sides form a triangle, determine whether it is equilateral, isosceles, or scalene

```python
side1 = input('Enter the side1: ')
side2 = input('Enter the side2: ')
side3 = input('Enter the side3: ')

if side1 + side2 > side3 and side2 + side1 > side3 and side3 + side1 > side2:
  if side1 == side2 and side1 == side3:
    print("Equilateral Triangle")
  elif side1 == side2 or side1 == side3 or side3 == side2:
    print("Isosceles Traingle")
  else:
    print("Scalene Triangle")
else:
  print("Not a valid triangle")
```

#3. Take marks (0–100) and print the corresponding grade (A/B/C/D/F).

```python
marks = int(input("Enter the marks: "))
if 0 > marks or marks > 100:
  print("invalid marks")
elif marks >=  90:
  print("A")
elif marks >= 80:
  print("B")
elif marks >= 70:
  print('C')
elif marks >= 35:
  print("D")
else:
  print("F")
```

#5.Take the hour of the day (0–23) and print “Good Morning”, “Good Afternoon”, “Good Evening”, or “Good Night”.

```python
hour = int(input("Enter the Hour: "))
if 4 <= hour < 12:
  print("Good Morning")
elif 12 <= hour < 17:
  print("Good Afternoon")
elif 17 <= hour < 21:
  print("Good Evening")
elif (21 <= hour <= 23) or (0 <= hour < 4):
  print("Good Night")
else:
  print("Invalid hour")
```

#8. Take an alphabet character and check if it lies between ‘a’ and ‘m’ or ‘n’ and ‘z’.

```python
char = input("Enter an alphabet character: ").lower()

if len(char) == 1 and char.isalpha():
  if 'a' <= char <= 'm':
    print(f'{char} lies between a and m')
  else:
    print(f'{char} lies between n and z')
else:
  print("Enter only single alphabet")
```
