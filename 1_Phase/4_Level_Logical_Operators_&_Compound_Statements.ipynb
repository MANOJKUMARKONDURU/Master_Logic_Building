#1. Take a character and check if it is a letter, a digit, or neither.

```python
def is_n_d(ch):
    if len(ch) == 1:
      if ch.isalpha():
        return "A Letter"
      elif ch.isdigit():
        return "A Digit"
      else:
        return "Neither"
    else:
      return "Enter only one character"

ch = input("Enter a character: ")
print(is_n_d(ch))
```

#2. Take a number and print “Fizz” if divisible by 3, “Buzz” if divisible by 5, and “FizzBuzz” if divisible by both.

```python
def is_divisible(num):
  if num % 3 == 0 and num % 5== 0:
    return "FizzBuzz"
  elif num % 3 == 0:
    return "Fizz"
  elif num % 5 == 0:
    return "Buzz"
  else:
    return "neither"

num = int(input("Enter a number: "))
print(is_divisible(num))
```

# 3. Take three numbers and print the median value (neither maximum nor minimum).

```python
def median(n1, n2, n3):
  if (n2 > n1 and n1 > n3) or (n3 > n1 and n1 > n2):
    return 'Number 1 is Median'
  elif (n1 > n2 and n2 > n3) or (n3 > n2 and n2 > n1):
    return 'Number 2 is Median'
  else:
    return 'Number 3 is Median'
n1 = int(input("Enter the number1: "))
n2 = int(input("Enter the number2: "))
n3 = int(input("Enter the number3: "))
print(median(n1, n2, n3))
```

#4. Take 24-hour time (hours and minutes) and print whether it is AM or PM.

```python
def is_am_pm(hour, minutes):
  if (hour >= 0 and hour <= 24) and (minutes >= 0 and minutes <= 60):
   return "AM" if 0 <= hour < 12 else "PM"
  else:
    return "Enter the correct time"

hour = int(input("Enter the hour: "))
minutes = int(input('enter the minutes'))
print(is_am_pm(hour, minutes))
```

#7. Take a single digit (0–9) and print its word form (“Zero” to “Nine”).

```python
def digit(num):
  if num == 0:
    return 'Zero'
  elif num == 1:
    return 'One'
  elif num == 2:
    return 'Two'
  elif num == 3:
    return 'Three'
  elif num == 4:
    return 'Four'
  elif num == 5:
    return 'Five'
  elif num == 6:
    return 'Six'
  elif num == 7:
    return 'Seven'
  elif num == 8:
    return 'Eight'
  elif num == 9:
    return 'Nine'
  else:
    return 'Enter number 0-9'

num = int(input('Enter a digit: '))
print(digit(num))
```

#9. Take electricity units consumed and calculate the bill as per slabs (using if-else).

```python
def electricity(units):
  bill = 0
  if units > 300:
    bill += (units - 300) * 35  #rate 35
    units = 300
  if units > 200:
    bill += (units - 200)* 30  #rate 35
    units = 200
  if units > 100:
    bill += (units - 100) * 25 #rate 25
    units = 100

  bill += units * 20

  return bill

units = int(input("Enter the number units consumed"))
print(electricity(units))
```

#10. Take a password string and check basic rules (length ≥ 8 and contains at least one digit)

```python
import re
def pwd_check(password):
  if len(password) >= 8:
    if re.search(r'\d', password):
      print("Yes, Valid Password")
    else:
      print("Not a valid password")
  else:
    print("Too short")

password = input("Enter the password: ")
pwd_check(password)
```
