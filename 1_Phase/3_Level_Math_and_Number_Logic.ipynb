#1. Take a 3-digit number and check if all digits are distinct.

```python
num = 321
h = num // 100
t = (num // 10) %10
o = (num) % 10

if h != t and t != o and o != h:
  print("All digits are distinct")
else:
  print("Not distinct")

if len(num) == 3 and len(set(num)) == 3:
    print("All digits are distinct")
else:
  print("Not distinct")
```

#2. Take a 3-digit number and determine if the middle digit is the largest, smallest, or neither.

```python
num = 213

h = num // 100
t = (num // 10) % 10
o = num % 10

if t > h and t > o:
  print('Largest')
elif t < h and t < o:
  print('Smallest')
else:
  print('neither')
```

#4. Check whether a given integer is single-digit, double-digit, or multi-digit.

```python
num = abs(int(input("Enter a number: ")))

if num < 10 :
  print("single-digit")
elif num < 100:
  print('double digit')
else:
  print('multi-digit')
```

#6. Take coordinates (x, y) and determine which quadrant the point lies in.

```python
x = int(input("Enter x value: "))
y = int(input("Enter y value: "))
if x > 0 and y > 0:
  print('Quadrant 1')
elif x < 0 and y > 0:
  print('Quadrant 2')
elif x < 0 and y < 0:
  print('Quadrant 3')
elif x > 0 and y < 0:
  print('Quadrant 4')
elif x == 0 and y == 0:
  print('origin')
else:
  print('on axis')
```

#10. Check whether a number is a perfect square (without using the square root function).

```python
num = int(input("Enter any number"))
i = 1
while i * i <= num:
  if i * i == num:
    print("perfect square")
    break
  i +=1
else:
  print("Not a perfect square")
```
